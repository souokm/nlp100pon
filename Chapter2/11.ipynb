```python
# タブ1文字につきスペース1文字に置換せよ．確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ
```

```python
import codecs
```

```python
with codecs.open('hightemp.txt','r','utf-8') as f:
    str = f.read()
    print(str)
    print("============replace()==============")
    print(str.replace("\t"," "))
```

```python
%%bash
sed -e s/$'\t'/" "/g hightemp.txt
```

```python
%%bash
cat hightemp.txt | tr "\t" " "
```

```python
%%bash
expand -t 1 hightemp.txt
```
